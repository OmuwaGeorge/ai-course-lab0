```python
import numpy as np
np.random.seed(42)
randomData = np.random.randn(100,3)

mean = randomData.mean(axis=0)
std = randomData.std(axis=0)

normalizeData = (randomData - mean)/std

print("Mean: ", normalizeData.mean(axis=0))
print("Std: ", normalizeData.std(axis=0))
```

```python
import numpy as np

X = np.random.randn(50, 3)
true_theta = np.array([2.5, -1.2, 3.7])
y = X @ true_theta + np.random.randn(50) * 0.1

X_b = np.c_[np.ones((X.shape[0], 1)), X]

true_theta_b = np.r_[0, true_theta]

theta_hat = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

print("Estimated coefficients:", theta_hat)
print("True theta:", true_theta_b)
```
